```python
# 데이터 처리(pandas), 수치 계산(numpy), 구글 드라이브 연동을 위한 라이브러리 import
import pandas as pd
import numpy as np
from google.colab import drive

drive.mount('/content/drive') # 구글 드라이브를 Colab에 연결(마운트)

path = '/content/drive/MyDrive/wine.csv'
```

```python
df = pd.read_csv(path)
df
```

```python
df2=df.drop('Wine',axis=1)
df2.head()
```

```python
y=df2['Alcohol']
y.value_counts()
y
```

```python
X=df2.drop('Alcohol',axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
model = LinearRegression()       # 선형 회귀 모델 생성
model.fit(X_train, y_train)      # 학습 데이터로 모델 학습
ly_preds = model.predict(X_test) # 테스트 데이터로 예측값 생성

print('평균제곱근오차', mean_squared_error(ly_preds, y_test))
# sklearn으로 MSE(평균제곱오차) 계산 및 출력

def mse_np(actual, predicted):
    return np.mean((np.array(actual) - np.array(predicted)) ** 2)
# numpy를 이용해 MSE 계산하는 함수 정의

print('평균제곱근오차', mse_np(ly_preds, y_test))
# numpy로 계산한 MSE 출력

def mse(actual, predicted):
    sum_square_error = sum((a - p) ** 2 for a, p in zip(actual, predicted))
    # 실제값과 예측값 차이를 제곱해서 모두 더함
    mean_square_error = sum_square_error / len(actual)
    # 전체 개수로 나눠 평균 계산
    return mean_square_error

print('평균제곱근오차', mse(ly_preds, y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

dy_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(dy_preds, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

ry_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(ry_preds, y_test))
```
